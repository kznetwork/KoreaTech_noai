```python
# load_digits를 활용한 train_test_split 실습

import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

# 1. 손글씨 데이터셋 로드
digits = load_digits()

# 독립변수 X (64개의 픽셀 값을 가진 이미지 데이터)
X = digits.data
# 종속변수 y (0부터 9까지의 진짜 숫자 정답 라벨)
y = digits.target

print("--- [1. 원본 데이터 세트 확인] ---")
print(f"전체 특성(X) 데이터 모양: {X.shape}  -> 1797개 샘플, 64개 변수")
print(f"전체 정답(y) 데이터 모양: {y.shape}  -> 1797개 정답")
print("-" * 60)
```

```python
# 2. train_test_split을 이용한 데이터 분할
# [주요 옵션 설명]
# - test_size=0.25 : 전체의 25%를 검증(Test)용으로, 75%를 학습(Train)용으로 분할
# - random_state=42 : 무작위 추출 결과를 고정하여 실행할 때마다 매번 똑같이 쪼개지도록 함
# - stratify=y : 0부터 9까지의 숫자가 균등한 비율로 Train과 Test에 들어가도록 계층적 분할 진행
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.25, 
    random_state=42, 
    stratify=y
)

# 3. 분할된 결과 구조 확인
print("--- [2. train_test_split 분할 결과 확인] ---")
print(f"학습용 특성 데이터 (X_train) 모양: {X_train.shape}")
print(f"테스트 특성 데이터 (X_test) 모양 : {X_test.shape}")
print(f"학습용 정답 데이터 (y_train) 모양: {y_train.shape}")
print(f"테스트 정답 데이터 (y_test) 모양 : {y_test.shape}")
print("-" * 60)
```

```python
# 4. stratify=y 옵션이 잘 작동했는지 비율 검증 (클래스별 개수)
print("--- [3. 정답 라벨(0~9) 비율 유지 확인] ---")
print("원본 데이터의 숫자별 빈도:\n", np.bincount(y))
print("테스트 데이터의 숫자별 빈도 (약 25%씩 균등 분할):\n", np.bincount(y_test))
```
