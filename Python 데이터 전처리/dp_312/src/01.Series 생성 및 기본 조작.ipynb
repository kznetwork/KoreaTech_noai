```python
# 1. Pandas 라이브러리 불러오기
# Series를 사용하려면 먼저 pandas 라이브러리를 가져와야 합니다. 통상적으로 pd라는 별칭으로 사용합니다.

import pandas as pd
import numpy as np  # 결측치(NaN) 표현을 위해 함께 불러옵니다.
```

```python
# 2. Series 생성하기
# Series는 1차원 배열 형태의 데이터 구조이며, 명시적인 인덱스(Index)를 가집니다.

# 방법 A: 리스트(List)로 생성 (인덱스는 자동으로 0, 1, 2... 지정)
data_list = [10, 20, 30, 40, 50]
s1 = pd.Series(data_list)
print("1. 리스트로 생성:")
print(s1)
print("-" * 30)

# 방법 B: 딕셔너리(Dictionary)로 생성 (Key가 인덱스가 됨)
data_dict = {'apple': 1500, 'banana': 3000, 'cherry': 5000}
s2 = pd.Series(data_dict)
print("2. 딕셔너리로 생성:")
print(s2)
print("-" * 30)

# 방법 C: 인덱스를 직접 지정하며 생성
s3 = pd.Series([90, 85, 95], index=['Math', 'Science', 'English'])
print("3. 인덱스 지정 생성:")
print(s3)
```

```python
# 3. Series 기본 속성 확인
# 생성한 Series의 구조와 정보를 파악할 때 사용하는 속성들입니다.

print(f"데이터 값 확인 (values): {s3.values}")
print(f"인덱스 확인 (index): {s3.index}")
print(f"데이터 타입 확인 (dtype): {s3.dtype}")
print(f"데이터 개수 확인 (size): {s3.size}")
```

```python
# 4. 데이터 선택 및 인덱싱 (Indexing & Slicing)

# 라벨을 통한 접근
print("수학 점수:", s3.loc['Math'])

# 위치(숫자)를 통한 접근
print("첫 번째 과목 점수:", s3.iloc[0])

# 여러 개 선택
print("수학, 영어 선택:\n", s3.loc[['Math', 'English']])

# 슬라이싱
# 라벨 슬라이싱은 끝값을 포함합니다.
print("Math부터 Science까지:\n", s3.loc['Math':'Science'])
```

```python
# 5. 조건 필터링 (Boolean Indexing)
# 조건에 맞는 데이터만 쏙 골라내는 아주 강력한 기능입니다.

# 90점 이상인 데이터만 필터링
over_90 = s3[s3 >= 90]
print("90점 이상인 과목:\n", over_90)
```

```python
# 6. Series 연산 및 기본 통계
# Pandas의 큰 장점은 모든 요소에 일괄적으로 연산을 적용(브로드캐스팅)할 수 있다는 점입니다.

# 일괄 연산 (모든 점수에 5점 더하기)
bonus_score = s3 + 5
print("보너스 점수 반영:\n", bonus_score)

# 주요 통계 메서드
print(f"평균 점수: {s3.mean()}")
print(f"최고 점수: {s3.max()}")
print(f"점수 요약 통계 정보:\n{s3.describe()}")
```

```python
# 7. 결측치(NaN) 처리하기
# 데이터 분석 시 자주 만나는 비어있는 값(NaN)을 다루는 방법입니다.

# 결측치를 포함한 Series 생성
s_na = pd.Series([1, np.nan, 3, np.nan, 5], index=['a', 'b', 'c', 'd', 'e'])

print("결측치 여부 확인 (True/False):\n", s_na.isna())

# 결측치를 다른 값(예: 0)으로 채우기
s_filled = s_na.fillna(0)
print("\n결측치를 0으로 채운 결과:\n", s_filled)

# 결측치가 있는 행 제거하기
s_dropped = s_na.dropna()
print("\n결측치를 제거한 결과:\n", s_dropped)
```
