```python
import pandas as pd

# 예시 데이터프레임
df = pd.DataFrame({
    '고객명': ['홍길동', '김철수', None],
    '생년': [1985, 2020, 1890],
    '이메일': ['hong@gmail.com', 'not-an-email', 'kim@naver.com']
})

# 평가 예시
print("결측치 확인:\n", df.isnull().sum())
print("생년 유효성 확인:\n", df[(df['생년'] < 1900) | (df['생년'] > 2024)])
print("이메일 유효성 확인:\n", df[~df['이메일'].str.contains('@')])
```
