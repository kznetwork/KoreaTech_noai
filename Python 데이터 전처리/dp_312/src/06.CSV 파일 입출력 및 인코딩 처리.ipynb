```python
# 1. Pandas 라이브러리 불러오기 및 실습용 데이터 준비
# 먼저 CSV 파일로 저장할 간단한 샘플 데이터프레임을 생성합니다. 한글 처리를 확인하기 위해 한글 데이터를 포함했습니다.

import pandas as pd
import numpy as np

# 실습용 샘플 데이터 생성 (한글 포함)
data = {
    '상품명': ['노트북', '스마트폰', '태블릿', '무선이어폰'],
    '가격': [1200000, 850000, 600000, 150000],
    '재고': [15, 30, 22, np.nan],  # 결측치(NaN) 포함
    '출시일': ['2025-01-10', '2025-03-15', '2025-05-20', '2026-02-01']
}
df = pd.DataFrame(data)
print("원본 데이터프레임:")
print(df)
```

```python
# 2. CSV 파일로 저장하기 (Output)
# 데이터프레임을 CSV 파일로 저장할 때는 to_csv() 메서드를 사용합니다.

# 방법 A: 가장 기본으로 저장 (인덱스 포함, UTF-8 인코딩)
df.to_csv('datasets/products_default.csv', encoding='utf-8')

# 방법 B: 실무에서 주로 사용하는 옵션 적용하여 저장
# - index=False: 무의미한 숫자 인덱스(0, 1, 2...)가 파일에 저장되는 것을 방지
# - na_rep='N/A': 결측치(NaN)를 빈칸 대신 'N/A'라는 문자로 채워서 저장
df.to_csv('datasets/products_clean.csv', index=False, na_rep='N/A', encoding='utf-8')

print("CSV 파일 저장 완료!")
```

```python
# 3. CSV 파일 읽어오기 (Input)
# 저장된 CSV 파일을 파이썬으로 불러올 때는 read_csv() 메서드를 사용합니다.

# 깨끗하게 저장한 CSV 파일 불러오기
# - keep_default_na=True (기본값): 'N/A'로 저장했던 결측치를 자동으로 다시 NaN(Null)으로 인식합니다.
df_read = pd.read_csv('datasets/products_clean.csv')

print("불러온 CSV 데이터 확인:")
print(df_read)
print("-" * 40)
print("데이터 정보(결측치 재인식 확인):")
df_read.info()
```

```python
# 4. [중요] 한글 깨짐 및 인코딩 오류 해결 가이드
# 한국어 환경에서 가장 자주 발생하는 오류는 UnicodeDecodeError입니다. 파일이 작성된 환경에 따라 인코딩 방식이 다르기 때문입니다.

# 상황 1: 엑셀(Excel)에서 만든 CSV 파일이 안 열릴 때
# 윈도우 환경의 엑셀에서 만든 CSV 파일이나, 엑셀에서 바로 열었을 때 한글이 깨지지 않게 하려면 cp949 또는 euc-kr 인코증을 사용해야 합니다.

# 엑셀 호환용으로 저장하기 (인코딩을 cp949로 지정)
df.to_csv('datasets/products_excel.csv', index=False, encoding='cp949')

# cp949로 저장된 파일 읽기 (encoding 옵션을 생략하면 에러가 날 수 있습니다)
df_excel = pd.read_csv('datasets/products_excel.csv', encoding='cp949')
print("Excel 인코딩(cp949) 파일 읽기 성공:")
print(df_excel.head(2))

# 상황 2: UTF-8인데 엑셀에서 한글이 깨져 보일 때 (BOM 처리)
# 인코딩을 utf-8로 저장한 CSV 파일을 윈도우 엑셀에서 그대로 더블클릭해 열면 한글이 심하게 깨집니다. 이때는 utf-8-sig 인코딩을 사용하면 엑셀이 파일의 시작 부분에서 UTF-8임을 감지하여 한글을 올바르게 표시합니다.

# UTF-8 BOM(Byte Order Mark)을 적용하여 저장
df.to_csv('datasets/products_utf8_bom.csv', index=False, encoding='utf-8-sig')
print("Excel에서 바로 더블클릭해 열어도 안 깨지는 UTF-8 파일 저장 완료!")
```
