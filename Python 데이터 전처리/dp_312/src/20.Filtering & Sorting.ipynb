```python
# 1. 라이브러리 불러오기 및 실습 데이터 생성
# 실습을 위해 간단한 전자제품 매장의 가상 재고 데이터를 생성합니다.

import pandas as pd

# 실습용 데이터 세트 생성
data = {
    '제품명': ['노트북', '스마트폰', '태블릿', '모니터', '무선이어폰', '기계식키보드'],
    '카테고리': ['가전', '모바일', '모바일', '가전', '모바일', '가전'],
    '가격': [1200000, 950000, 700000, 350000, 180000, 120000],
    '재고': [15, 8, 22, 5, 40, 0],
    '평점': [4.5, 4.8, 4.3, 4.2, 4.6, 4.7]
}

df = pd.DataFrame(data)
print("--- 원본 데이터프레임 ---")
print(df)
```

```python
# 2. 데이터 필터링 (Filtering)
# 특정 조건을 만족하는 행(Row)만 추출하는 방법입니다.

# ① 단일 조건 필터링
# 가격이 500,000원 이상인 제품만 필터링
expensive_products = df[df['가격'] >= 500000]
print("1. 가격이 50만 원 이상인 제품:")
print(expensive_products[['제품명', '가격']])
print("-" * 50)

# 재고가 0인(품절된) 제품만 필터링
sold_out = df[df['재고'] == 0]
print("2. 품절된 제품:")
print(sold_out)
```

```python
# ② 다중 조건 필터링 (And, Or)
# •	And 연산 (&): 두 조건을 모두 만족해야 함

# •	Or 연산 (|): 두 조건 중 하나만 만족해도 됨

# 주의: 각 조건은 반드시 괄호 ()로 묶어주어야 에러가 나지 않습니다.

# And 조건: 카테고리가 '모바일'이면서 평점이 4.5 이상인 제품
cond_and = (df['카테고리'] == '모바일') & (df['평점'] >= 4.5)
print("3. 모바일 기기 중 평점 4.5 이상:")
print(df[cond_and])
print("-" * 50)

# Or 조건: 가격이 1,000,000원 이상이거나 재고가 5개 이하인 제품
cond_or = (df['가격'] >= 1000000) | (df['재고'] <= 5)
print("4. 고가 제품 또는 재고 부족 제품:")
print(df[cond_or])
```

```python
# ③ 문자열 및 목록 포함 필터링 (isin())
# 여러 특정 값 중 하나라도 포함되어 있는지 확인할 때 유용합니다.

# 카테고리가 '가전' 또는 '사무'인 데이터 찾기 (여기서는 가전만 해당)
target_categories = ['가전', '사무']
print("5. 특정 카테고리 목록에 포함된 제품:")
print(df[df['카테고리'].isin(target_categories)])
```

```python
# 3. 데이터 정렬 (Sorting)
# 특정 기준에 따라 데이터의 순서를 재배치하는 방법입니다.

# ① 단일 컬럼 기준 값 정렬 (sort_values)
# 가격 기준 오름차순 정렬 (기본값)
df_price_asc = df.sort_values(by='가격')
print("1. 가격 기준 오름차순 정렬:")
print(df_price_asc[['제품명', '가격']])
print("-" * 50)

# 평점 기준 내림차순 정렬 (ascending=False)
df_rating_desc = df.sort_values(by='평점', ascending=False)
print("2. 평점 기준 내림차순 정렬:")
print(df_rating_desc[['제품명', '평점']])
```

```python
# ② 다중 컬럼 기준 값 정렬
# 첫 번째 정렬 기준의 값이 같을 때, 두 번째 컬럼을 기준으로 삼아 정렬합니다.

# 카테고리 기준으로 오름차순 정렬하되, 같은 카테고리 내에서는 가격 기준 내림차순 정렬
# by와 ascending 옵션에 각각 리스트를 전달합니다.
df_multi_sort = df.sort_values(by=['카테고리', '가격'], ascending=[True, False])
print("3. 카테고리(오름차순) -> 가격(내림차순) 정렬:")
print(df_multi_sort)
```

```python
# ③ 인덱스 기준 정렬 (sort_index)
# 값이 아니라 왼쪽의 행 인덱스(0, 1, 2...) 번호 순서대로 다시 정렬하고 싶을 때 사용합니다. 정렬을 거치며 뒤죽박죽된 데이터프레임을 원상복구할 때 주로 씁니다.

# 예시를 위해 가격 정렬된 데이터프레임 사용
print("4. 원래 인덱스 순서대로 역정렬(내림차순):")
print(df_price_asc.sort_index(ascending=False))
```

```python
# ※ Filtering & Sorting 응용 조합
# 실무에서는 보통 필터링을 먼저 거친 후 정렬까지 이어서 수행하는 경우가 많습니다.

# 시나리오: 가전 카테고리 제품 중에서, 재고가 남아있는 제품만 골라 가격이 저렴한 순서대로 정렬하기
result = df[(df['카테고리'] == '가전') & (df['재고'] > 0)].sort_values(by='가격', ascending=True)

print("--- [응용 결과] 살 수 있는 가전제품 중 가장 저렴한 순 ---")
print(result)
```
