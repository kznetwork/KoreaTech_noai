```python
import pandas as pd

# 샘플 데이터 생성 (학생 성적표)
data = {
    '이름': ['김철수', '이영희', '박민수', '최지우'],
    '국어': [90, 85, 95, 70],
    '영어': [85, 100, 80, 90],
    '수학': [95, 90, 85, 85]
}

df = pd.DataFrame(data)
# 행 구분을 쉽게 하기 위해 이름을 '인덱스'로 지정합니다.
df.set_index('이름', inplace=True)

print("-- 원본 데이터프레임 --")
print(df)
```

```python
# 1. 열(Column) 데이터 가져오기
# 열을 가져올 때는 대괄호 [ ] 안에 열 이름을 적어주면 됩니다.

# '국어' 열 하나만 가져오기
kor_scores = df['국어']
print(kor_scores)

# 여러 개의 열('국어', '수학') 한 번에 가져오기 (대괄호를 두 번 사용합니다)
sub_scores = df[['국어', '수학']]
print(sub_scores)
```

```python
# 2. 행(Row) 데이터 가져오기
# 행을 가져올 때는 보통 두 가지 방법을 사용합니다.
# ∧	loc: 사람이 읽을 수 있는 행 이름(인덱스)으로 가져올 때
# ∧	iloc: 컴퓨터가 인식하는 행 번호(0부터 시작하는 위치)로 가져올 때

# 방법 1: loc (행 이름으로 가져오기)
younghee_data = df.loc['이영희']
print(younghee_data)

# 방법 2: iloc (행 번호로 가져오기 - 첫 번째 행은 0번)
first_row = df.iloc[0]  # '김철수'의 데이터
print(first_row)

# 여러 행 가져오기 (슬라이싱 사용)
two_rows = df.iloc[0:2]  # 0번부터 1번 행까지 (2번은 제외)
print(two_rows)
```
