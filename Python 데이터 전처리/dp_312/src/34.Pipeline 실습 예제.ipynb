```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# 1. 손글씨 데이터셋 로드
digits = datasets.load_digits()
features = digits.data
target = digits.target

# 2. 훈련/테스트 세트 분리
# [보완] stratify=target을 추가하여 0~9까지의 정답 비율을 균등하게 쪼갭니다.
X_train, X_test, y_train, y_test = train_test_split(
    features, target, 
    test_size=0.1, 
    random_state=1,
    stratify=target
)

# 3. Pipeline 생성
# [보완] max_iter=2000을 추가하여 경고(ConvergenceWarning) 발생을 원천 차단합니다.
pipeline = make_pipeline(
    StandardScaler(), 
    LogisticRegression(max_iter=2000, random_state=1)
)

# 4. 파이프라인 모델 훈련
# (StandardScaler의 fit_transform과 LogisticRegression의 fit이 내부에서 연속 수행됨)
pipeline.fit(X_train, y_train)

# 5. 예측
y_pred = pipeline.predict(X_test)

# 6. 평가 및 결과 출력
# [보완] 임포트 구문을 최상단으로 이동시키고 결과 출력을 정돈했습니다.
print("=== 최종 모델 평가 결과 ===")
print(f"테스트 데이터 정확도: {accuracy_score(y_test, y_pred) * 100:.2f}%\n")

print("[상세 분류 리포트]")
print(classification_report(y_test, y_pred))
```
