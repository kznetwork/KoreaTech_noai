```python
import pandas as pd
import numpy as np

# 1. 함수 정의 (사용자가 작성한 코드)
def check_missing_col(dataframe):
    missing_col = []
    counted_missing_col = 0
    for i, col in enumerate(dataframe.columns):
        missing_values = sum(dataframe[col].isna())
        is_missing = True if missing_values >= 1 else False
        if is_missing:
            counted_missing_col += 1
            print(f'결측치가 있는 컬럼은: {col}입니다')
            print(f'해당 컬럼에 총 {missing_values}개의 결측치가 존재한다.')
            missing_col.append([col, dataframe[col].dtype])
    if counted_missing_col == 0:
        print('결측치가 존재하지 않다')
    return missing_col
```

```python
# 2. 'SW특기' 컬럼에 정확히 2개의 결측치(NaN)가 있는 실습용 데이터 생성
data = {
    '이름': ['김범수', '나얼', '박효신', '이수', '신용재'],
    '나이': [45, 47, 44, 45, 36],
    '소속사': ['영엔터', '롱플레이', '글러브', '325이엔씨', '밀리언마켓'],
    'SW특기': ['Python', np.nan, 'C++', np.nan, 'Java']  # 결측치 2개 배치
}

df = pd.DataFrame(data)

print("--- [실습에 사용될 데이터프레임] ---")
print(df)
print("-" * 50)
```

```python
# 3. 함수 실행 및 결과 출력
print("--- [함수 실행 결과 확인] ---")
missing_col = check_missing_col(df)


print("-" * 50)
print("--- [반환된 리스트 변수(missing_col) 확인] ---")
print(missing_col)
```
