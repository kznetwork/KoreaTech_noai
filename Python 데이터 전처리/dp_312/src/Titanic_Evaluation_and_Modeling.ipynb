```python
train['Pclass'].value_counts()
```

### 문제 09. 성별('Gender') 및 탑승 항구('Embarked')별 탑승객 분포(빈도수)를 확인하는 코드를 각각 작성하시오.

```python
print(train['Gender'].value_counts())
print(train['Embarked'].value_counts())
```

### 문제 10. 'Embarked' 필드의 결측치를 데이터 빈도수가 가장 많은 값인 'S'로 할당(대체)하는 코드를 작성하시오.

```python
train['Embarked'] = train['Embarked'].fillna('S')
```

### 문제 11. 'Age' 필드의 결측치를 해당 필드의 중간값(Median)으로 할당(대체)하는 코드를 작성하시오.

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
```

### 문제 12. 'Name' 필드에서 'Mr', 'Mrs', 'Miss'와 같이 신분이나 호칭을 나타내는 단어를 정규표현식으로 추출하여 'Title'이라는 새로운 파생변수 필드에 할당하는 코드를 작성하시오.

```python
train['Title'] = train['Name'].str.extract('([A-Za-z]+)\\.', expand=False)
```

### 문제 13. 'Title' 필드에 존재하는 다양한 호칭들을 아래의 기준에 맞게 6개의 대표 카테고리로 변경(치환)하는 코드를 작성하시오.
* *기준: ['Capt', 'Col', 'Major', 'Dr', 'Rev'] ➡️ 'Officer' / ['Jonkheer', 'Master'] ➡️ 'Master' / ['Don', 'Sir', 'the Countess', 'Lady', 'Dona'] ➡️ 'Royalty' / ['Mme', 'Ms', 'Mrs'] ➡️ 'Mrs' / ['Mlle', 'Miss'] ➡️ 'Miss' / ['Mr'] ➡️ 'Mr'*

```python
train['Title'] = train['Title'].replace(['Capt', 'Col', 'Major', 'Dr', 'Rev'], 'Officer')
train['Title'] = train['Title'].replace(['Jonkheer', 'Master'], 'Master')
train['Title'] = train['Title'].replace(['Don', 'Sir', 'the Countess', 'Lady', 'Dona'], 'Royalty')
train['Title'] = train['Title'].replace(['Mme', 'Ms', 'Mrs'], 'Mrs')
train['Title'] = train['Title'].replace(['Mlle', 'Miss'], 'Miss')
train['Title'] = train['Title'].replace('Mr', 'Mr')
```

### 문제 14. 연속형 변수인 나이('Age') 필드를 구간별로 그룹핑하여 'AgeGroup'이라는 범주형 파생필드를 생성하시오.
* *구간 및 레이블 기준: 0~18세('Baby'), 19~25세('Youth'), 26~35세('YoungAdult'), 36~60세('MiddleAged'), 61~100세('Senior')*

```python
bin = [0, 18, 25, 35, 60, 100]
group_names = ['Baby', 'Youth', 'YoungAdult', 'MiddleAged', 'Senior']
train['AgeGroup'] = pd.cut(train['Age'], bins=bin, labels=group_names)
```

## 03-03. 데이터 준비 및 피처 인코딩

### 문제 01. 범주형 변수를 정수형 타입으로 변환하기 위해 필요한 사이킷런의 `LabelEncoder` 모듈을 임포트하시오.

```python
from sklearn.preprocessing import LabelEncoder
```

### 문제 02. 머신러닝 모델 학습에 불필요하거나 과적합을 유발할 수 있는 'Name', 'Ticket', 'SibSp', 'Parch', 'Cabin' 컬럼들을 데이터프레임에서 삭제하는 코드를 작성하시오.

```python
train = train.drop(['Name', 'Ticket', 'SibSp', 'Parch', 'Cabin'], axis=1)
```

### 문제 03. 문자열 및 범주형 데이터인 ['Gender', 'Embarked', 'Title', 'AgeGroup'] 컬럼들을 반복문을 사용해 전부 문자열로 안전하게 변환한 후, `LabelEncoder`를 활용하여 정수형 데이터로 일괄 인코딩하는 코드를 작성하시오.

```python
label = LabelEncoder()
for col in ['Gender', 'Embarked', 'Title', 'AgeGroup']:
    train[col] = train[col].astype(str)
    train[col] = label.fit_transform(train[col])
```

### 문제 04. 학습에 사용될 독립변수(특성) 데이터프레임 `X_train`과 예측 목표인 종속변수(라벨) 데이터프레임 `Y_train`으로 분리하는 코드를 작성하시오.

```python
X_train = train[['PassengerId', 'Pclass', 'Gender', 'Age', 'Fare', 'Embarked', 'Title', 'AgeGroup']]
Y_train = train['Survived']
```

## 03-04. 모델 생성 및 예측 (Logistic Regression)

### 문제 01. 로지스틱 회귀(Logistic Regression) 모델을 사용하기 위해 필요한 모듈을 임포트하고, 모델 객체를 생성하여 변수 `lr`에 할당하시오.

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
```

### 문제 02. 일반화 성능 평가를 위한 테스트 데이터셋인 `test.csv` 파일을 읽어와 변수 `test`에 로드하는 코드를 작성하시오.

```python
test = pd.read_csv('../datasets/titanic/test.csv')
```

### 문제 03. 불러온 `test` 데이터셋에 결측치(Null)가 존재하는지 컬럼별로 확인하는 코드를 작성하시오.

```python
test.isnull().sum()
```

### 문제 04. `test` 데이터셋의 결측치 처리를 위해 'Age' 컬럼은 중앙값으로, 'Fare' 컬럼은 평균값으로 결측치를 대체하고, 'Embarked' 컬럼은 최빈값인 'S'로 대체하는 코드를 작성하시오.

```python
test['Age'] = test['Age'].fillna(test['Age'].median())
test['Fare'] = test['Fare'].fillna(test['Fare'].mean())
test['Embarked'] = test['Embarked'].fillna('S')
```

### 문제 05. **[수험생 유의사항]** 모델이 정상적으로 예측을 수행할 수 있도록, `test` 데이터셋도 `train` 데이터셋과 완벽히 동일한 컬럼 구조와 인코딩 모델을 갖추도록 파생변수 생성 및 피처 가공 작업을 수행하고 최종 독립변수 데이터프레임 `X_test`를 정의하시오.

```python
# 1. test 데이터셋의 호칭(Title) 파생변수 생성 및 매핑
test['Title'] = test['Name'].str.extract('([A-Za-z]+)\\.', expand=False)
test['Title'] = test['Title'].replace(['Capt', 'Col', 'Major', 'Dr', 'Rev'], 'Officer')
test['Title'] = test['Title'].replace(['Jonkheer', 'Master'], 'Master')
test['Title'] = test['Title'].replace(['Don', 'Sir', 'the Countess', 'Lady', 'Dona'], 'Royalty')
test['Title'] = test['Title'].replace(['Mme', 'Ms', 'Mrs'], 'Mrs')
test['Title'] = test['Title'].replace(['Mlle', 'Miss'], 'Miss')
test['Title'] = test['Title'].replace('Mr', 'Mr')
test['Title'] = test['Title'].fillna('Mr')

# 2. test 데이터셋의 연령대(AgeGroup) 파생변수 생성
bin = [0, 18, 25, 35, 60, 100]
group_names = ['Baby', 'Youth', 'YoungAdult', 'MiddleAged', 'Senior']
test['AgeGroup'] = pd.cut(test['Age'], bins=bin, labels=group_names)

# 3. 범주형 변수 일괄 라벨 인코딩 동기화 적용
for col in ['Gender', 'Embarked', 'Title', 'AgeGroup']:
    test[col] = test[col].astype(str)
    test[col] = label.fit_transform(test[col])

# 4. train 데이터셋과 차원(컬럼 세트 및 순서) 일치화
X_test = test[['PassengerId', 'Pclass', 'Gender', 'Age', 'Fare', 'Embarked', 'Title', 'AgeGroup']]
```

### 06. 앞서 정의한 로지스틱 회귀 객체 `lr`에 `X_train`과 `Y_train` 데이터를 입력하여 모델을 학습(훈련)시키고, 가공된 테스트 데이터 `X_test`를 투입하여 예측결과를 도출한 뒤 변수 `y_pred`에 저장하시오.

```python
# 모델 학습 수행
lr.fit(X_train, Y_train)

# 검증 데이터 기반 예측 수행
y_pred = lr.predict(X_test)
```
