```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# 1. 데이터셋 로드
digits = load_digits()
X = digits.data
y = digits.target

print("--- [1. 데이터 준비] ---")
print(f"특성 데이터(X) 모양: {X.shape}")
print(f"정답 데이터(y) 모양: {y.shape}\n")
```

```python
# 2. 파이프라인(Pipeline) 구축
# 전처리(StandardScaler)와 모델 학습(LogisticRegression) 단계를 하나로 묶습니다.
# 이렇게 묶어두면 K-Fold 내부에서 Train 폴드로만 스케일러를 학습(fit)시키고 
# Validation 폴드는 변환(transform)만 하도록 자동 제어되어 데이터 누수가 방지됩니다.
pipeline = Pipeline([
    ('scaler', StandardScaler()),                 # 1단계: 데이터 표준화 전처리
    ('classifier', LogisticRegression(max_iter=2000, random_state=42)) # 2단계: 분류 모델 설정
])
```

```python
# 3. K-Fold 교차 검증 설정
# 데이터를 총 5개의 덩어리(Fold)로 나눕니다. (5-Fold 교차 검증)
# shuffle=True는 데이터를 나누기 전에 순서를 한 번 섞어주는 옵션입니다.
kf = KFold(n_splits=5, shuffle=True, random_state=42)


# 4. 교차 검증 수행 및 평가
# cross_val_score 함수가 파이프라인과 K-Fold 설정을 입력받아 자동으로 5번의 학습과 검증을 반복합니다.
print("--- [2. 5-Fold 교차 검증 시작] ---")
scores = cross_val_score(pipeline, X, y, cv=kf, scoring='accuracy', n_jobs=-1)
```

```python
# 5. 최종 결과 출력
# 5번 반복할 때마다 나온 각각의 정확도와, 이들의 평균 정확도를 확인합니다.
print("[각 폴드별 예측 정확도(Accuracy)]")
for i, score in enumerate(scores):
    print(f"Fold {i+1}: {score*100:.2f}%")

print("-" * 40)
print(f"💡최종 평균 정확도: {np.mean(scores)*100:.2f}%")
```
