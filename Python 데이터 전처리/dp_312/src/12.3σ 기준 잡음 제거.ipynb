```python
import numpy as np
import pandas as pd

data = [10, 12, 11, 999, 13, 14]   # 999는 잡음

mean = np.mean(data)    # 176.5
std  = np.std(data)     # 363.5

# 3σ 이상 제거
clean_data = [x for x in data if abs(x - mean) <= 3 * std]
print(clean_data)  # [10, 12, 11, 13, 14]

# pandas 버전
s = pd.Series(data)
z_score = (s - s.mean()) / s.std()
clean_s = s[abs(z_score) <= 3]
print(clean_s.tolist())  # [10, 12, 11, 13, 14]
```

```python
import pandas as pd

data = [10, 12, 11, 999, 13, 14]

s = pd.Series(data)

q1 = s.quantile(0.25)
q3 = s.quantile(0.75)

iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

clean_s = s[(s >= lower) & (s <= upper)]

print(clean_s.tolist())
```
