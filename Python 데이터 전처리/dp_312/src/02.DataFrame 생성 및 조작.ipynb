```python
# 1. Pandas 라이브러리 불러오기

import pandas as pd
import numpy as np
```

```python
# 2. DataFrame 생성하기
# DataFrame은 행(Row)과 열(Column)로 구성되며, 딕셔너리나 리스트 등 다양한 데이터 형태로 생성할 수 있습니다.

# 방법 A: 딕셔너리(Dictionary)를 이용한 생성 (가장 많이 사용됨)
# Key는 열(Column) 이름이 되고, Value(리스트)는 해당 열의 데이터가 됩니다.
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 35],
    'City': ['Seoul', 'Busan', 'Incheon', 'Daegu'],
    'Score': [85, 92, 78, 95]
}
df = pd.DataFrame(data_dict)
print("1. 딕셔너리로 생성한 DataFrame:")
print(df)
print("-" * 40)

# 방법 B: 중첩 리스트와 컬럼명 지정을 통한 생성
data_list = [
    ['Alice', 25, 'Seoul', 85],
    ['Bob', 30, 'Busan', 92],
    ['Charlie', 22, 'Incheon', 78],
    ['David', 35, 'Daegu', 95]
]
df_list = pd.DataFrame(data_list, columns=['Name', 'Age', 'City', 'Score'])
print("2. 리스트로 생성한 DataFrame:")
print(df_list)
```

```python
# 3. 데이터 확인 및 탐색 (기본 메서드)
# 데이터프레임의 크기가 클 때, 전체 구조와 앞부분 데이터만 빠르게 확인할 때 사용합니다.

# 상위 2개 행만 확인 (기본값은 5개)
print("상위 데이터 확인 (head):")
print(df.head(2))
print("-" * 40)

# 데이터프레임의 기본 정보 확인 (데이터 타입, 결측치 아닌 개수 등)
print("DataFrame 정보 확인 (info):")
df.info()
print("-" * 40)

# 형태 확인 (행, 열 개수 반환)
print(f"데이터 모양 확인 (shape): {df.shape}")  # 결과: (4, 4) -> 4행 4열
```

```python
# 4. 열(Column)과 행(Row) 데이터 선택하기
# --- [열 선택] ---
# 단일 열 선택 (Series 형태로 반환됨)
print("Age 열 선택:")
print(df['Age'])
print("-" * 40)

# 여러 열 선택 (대괄호 두 개 사용, DataFrame 형태로 반환됨)
print("Name과 City 열 선택:")
print(df[['Name', 'City']])
print("-" * 40)


# --- [행 선택: loc와 iloc] ---
# loc: 인덱스 라벨(이름) 기준 선택
print("0번 이름의 행 선택 (loc):")
print(df.loc[0])
print("-" * 40)

# iloc: 컴퓨터 내부 정수 위치 기준 선택 (슬라이싱 가능)
print("0번부터 1번 행까지 선택 (iloc):")
print(df.iloc[0:2])
```

```python
# 5. 데이터 조건 필터링 (Boolean Indexing)
# 특정 조건을 만족하는 행만 추출하는 방법입니다.

# 나이가 25세 이상인 데이터만 추출
age_over_25 = df[df['Age'] >= 25]
print("나이가 25세 이상인 행:\n", age_over_25)
print("-" * 40)

# 여러 조건 결합 (And: &, Or: |) - 각 조건은 반드시 괄호()로 묶어야 합니다.
cond = (df['Age'] >= 25) & (df['Score'] >= 90)
filtered_df = df[cond]
print("나이가 25세 이상이고 점수가 90점 이상인 행:\n", filtered_df)
```

```python
# 6. 열 생성, 수정 및 삭제 (정렬 포함)
# 기존 데이터를 바탕으로 새 열을 만들거나 필요 없는 열을 지우는 작업입니다.

# 새 열 추가 (모두 동일한 값으로 채우기)
df['Pass'] = 'Yes'

# 기존 열을 활용한 계산으로 새 열 추가
df['Score_Bonus'] = df['Score'] + 5
print("열 추가 후 DataFrame:\n", df)
print("-" * 40)

# 데이터 정렬 (Score 기준 내림차순 정렬)
df_sorted = df.sort_values(by='Score', ascending=False)
print("Score 기준 내림차순 정렬:\n", df_sorted)
print("-" * 40)

# 열 삭제 (axis=1은 열을 의미, inplace=True를 하면 원본 df가 직접 수정됨)
df.drop('Pass', axis=1, inplace=True)
print("Pass 열 삭제 후 DataFrame:\n", df)
```

```python
# 7. 그룹화 및 통계 (Groupby)
# 특정 카테고리별로 묶어서 평균, 합계 등의 통계를 낼 때 사용합니다. (실습을 위해 간단한 부서 데이터를 새로 생성)

# 실습용 신규 데이터프레임 생성
company_data = {
    'Team': ['A', 'B', 'A', 'B', 'A'],
    'Sales': [100, 200, 150, 250, 300]
}
df_company = pd.DataFrame(company_data)

# Team별 Sales의 평균 구하기
team_avg = df_company.groupby('Team').mean()
print("팀별 매출 평균:")
print(team_avg)
```
