```python
# 1. f-string 방식 (가장 추천 / 파이썬 3.6 이상)
# 현재 파이썬 표준이자 가장 직관적이고 가독성이 좋은 방식입니다. 문자열 앞에 f 또는 F를 붙이고, 중괄호 {} 안에 변수나 표현식을 직접 넣습니다.

name = "민준"
age = 27
height = 178.5

# 기본 f-string
print(f"안녕하세요, 제 이름은 {name}이고 나이는 {age}세입니다.")

# 중괄호 안에서 간단한 연산도 가능
print(f"내년에는 {age + 1}세가 됩니다.")
print(f"키를 정수로 반환하면 {int(height)}cm입니다.")
```

```python
# 2. format() 메서드 방식 (파이썬 3 이상)
# 문자열 뒤에 .format()을 붙여서 값을 전달하는 방식입니다. f-string이 나오기 전까지 가장 널리 쓰였습니다.

# 방법 A: 순서대로 매핑 (중괄호 순서대로 format 안의 인자가 들어감)
print("이름: {}, 나이: {}".format(name, age))

# 방법 B: 인덱스 번호 지정 (순서를 바꾸거나 재사용할 때 유용)
print("나이: {1}, 이름: {0}, 다시 나이: {1}".format(name, age))

# 방법 C: 이름(Key) 지정
print("안녕하세요 {user}님, {location}에 오신 것을 환영합니다.".format(user="지은", location="서울"))
```

```python
# 3. % 서식 지정자 방식 (과거 방식 / 파이썬 2부터 존재)
# C언어 스타일의 가장 오래된 방식입니다. 데이터 타입에 따라 다른 기호를 써야 하므로 다소 번거롭지만, 레거시 코드에서 종종 볼 수 있습니다.

# -	%s: 문자열 (String)

# -	%d: 정수 (Integer)

# -	%f: 실수 (Floating-point)

# 변수 타입에 맞는 서식 지정자를 사용해야 합니다.
print("이름: %s, 나이: %d세, 키: %fcm" % (name, age, height))

# ※ 실무 필수: 소수점 및 공백 자리수 제어 (응용)
# 데이터 분석이나 보고서를 작성할 때 수치를 깔끔하게 정렬하는 포맷팅 기법입니다. (가장 편리한 f-string 기준으로 작성했습니다)

# ① 소수점 자리수 제한
# 실수 데이터의 지저분한 소수점 아래 자리를 깔끔하게 끊어줍니다.

pi = 3.1415926535

# 소수점 아래 두 자리까지만 출력 (.2f)
print(f"원주율: {pi:.2f}")  # 결과: 3.14
```

```python
# ② 천 단위 콤마(,) 표시
# 금액이나 통계 수치를 표현할 때 매우 유용합니다.

money = 5000000

# 숫자에 3자리마다 콤마 추가 (:,)
print(f"이번 달 매출은 {money:,}원입니다.")  # 결과: 5,000,000원
```

```python
# ③ 공백 정렬 및 자리수 채우기
# 문자열의 길이를 고정하고 정렬 상태를 맞출 때 사용합니다.

text = "파이썬"

# 10칸을 확보하고 왼쪽 정렬 (<), 오른쪽 정렬 (>), 가운데 정렬 (^)
print(f"'{text:<10}'")  # 결과: '파이썬       '
print(f"'{text:>10}'")  # 결과: '       파이썬'
print(f"'{text:^10}'")  # 결과: '   파이썬    '

# 빈 공간을 공백 대신 특정 문자(예: 0)로 채우기
number = 42
print(f"{number:05d}")  # 결과: 00042 (총 5칸을 확보하고 빈 곳은 0으로 채움)
```
