```python
# 1. Pandas 라이브러리 불러오기 및 실습 데이터 준비
# 한글과 숫자가 섞인 간단한 인사 및 매출 데이터를 생성합니다.

import pandas as pd
import numpy as np

# 1. 인사 정보 데이터프레임
data_employees = {
    '사원번호': ['E001', 'E002', 'E003', 'E004'],
    '이름': ['김철수', '이영희', '박민준', '최서연'],
    '부서': ['영업부', '기획부', '개발부', '영업부'],
    '급여': [3500000, 4000000, 4500000, np.nan] # 결측치 포함
}
df_emp = pd.DataFrame(data_employees)

# 2. 매출 정보 데이터프레임
data_sales = {
    '분기': ['1분기', '2분기', '3분기', '4분기'],
    '매출액': [15000000, 23000000, 18000000, 29000000]
}
df_sales = pd.DataFrame(data_sales)

print("준비된 인사 데이터:")
print(df_emp)
```

```python
# 2. 엑셀 파일로 저장하기 (Output)
# 기본적인 단일 시트 저장 방법과, 하나의 엑셀 파일 안에 여러 개의 시트(Sheet)를 나누어 저장하는 방법입니다.

# 방법 A: 단일 시트로 저장하기
# index=False 옵션으로 0, 1, 2... 와 같은 기본 인덱스가 저장되는 것을 방지합니다.
# sheet_name으로 시트의 이름을 지정할 수 있습니다.
df_emp.to_excel('datasets/employee_info.xlsx', index=False, sheet_name='인사정보')
print("단일 시트 엑셀 저장 완료!")

# 방법 B: 하나의 파일에 다중 시트(Multiple Sheets)로 저장하기
# 여러 개의 데이터프레임을 하나의 엑셀 파일 내 각각 다른 시트에 넣으려면 pd.ExcelWriter를 사용해야 합니다.

# ExcelWriter를 컨텍스트 매니저(with문)와 함께 사용
with pd.ExcelWriter('datasets/company_data.xlsx', engine='openpyxl') as writer:
    df_emp.to_excel(writer, sheet_name='사원현황', index=False)
    df_sales.to_excel(writer, sheet_name='매출현황', index=False)

print("다중 시트 엑셀 저장 완료!")
```

```python
# 3. 엑셀 파일 읽어오기 (Input)
# 저장된 엑셀 파일을 다시 파이썬으로 불러오는 방법입니다.

# 방법 A: 기본적으로 첫 번째 시트를 읽어옴
df_read_first = pd.read_excel('datasets/company_data.xlsx')
print("첫 번째 시트(사원현황) 불러오기 결과:")
print(df_read_first.head(2))
print("-" * 40)

# 방법 B: 특정 시트 이름을 지정하여 읽어옴
df_read_sales = pd.read_excel('datasets/company_data.xlsx', sheet_name='매출현황')
print("특정 시트(매출현황) 불러오기 결과:")
print(df_read_sales)
```
